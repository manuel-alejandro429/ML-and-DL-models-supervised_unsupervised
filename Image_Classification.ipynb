# Image classification with classical Machine Learning Strategics

```python
import numpy as np

import matplotlib.pyplot as plt

from keras.datasets import mnist

import sklearn.metrics as skm
from sklearn.metrics import accuracy_score
```

```python
# The MNIST database is loaded:
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# The size of the training and test data is printed:
print("x_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print("x_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)

# A random image of the training set is displayed:
rand_index = np.random.randint(0, x_train.shape[0])
plt.imshow(x_train[rand_index], cmap="gray")
plt.title("Label: {}".format(y_train[rand_index]))
plt.show()

# Training and validation images are vectorized:
x_train=np.reshape(x_train,(x_train.shape[0],x_train.shape[1]*x_train.shape[2]))
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1]*x_test.shape[2]))

# The training and validation set is normalized:
x_train=x_train/255
x_test=x_test/255
```

1. Logistic Regresion (LR):

```python
from sklearn.linear_model import LogisticRegression

# Training

# A logistic regression model is created and fitted with the training data:

lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
```

```python
# Validation

# The LR model is used to make prediction on the test data:
predictions = lr_model.predict(x_test)

# The performance of the LR model is evaluated:
accuracy = accuracy_score(y_test, predictions)
print("Accuracy LR:",accuracy)

# Confusion matrix, allows us to visually analyze the results:
cm = skm.confusion_matrix(y_test, predictions)
disp = skm.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=None)
disp.plot()
plt.show()
```

2. Decision Tree (DT):

```python
from sklearn.tree import DecisionTreeClassifier

# Training
# A decision tree (DT) model is created and fitted with the training data:

dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
```

```python
# Validation
#  The DT model is used to make prediction on the test data:

predictions = dt_model.predict(x_test)

# Evaluating the model

accuracy = accuracy_score(y_test, predictions)
print("Accuracy DT:",accuracy)

#Confusion matrix

cm = skm.confusion_matrix(y_test, predictions)
disp = skm.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=None)
disp.plot()
plt.show()
```

3. Random forest (RF):

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(x_train, y_train)
```

```python
# Evaluating the model
predictions = rf_model.predict(x_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy RF:",accuracy)


#Confusion matrix
cm = skm.confusion_matrix(y_test, predictions)
disp = skm.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=None)
disp.plot()
plt.show()
```

4. Naive Bayes (NB):

```python
from sklearn.naive_bayes import GaussianNB

# Training:

# A Naive Bayes (NB) model is created and fitted with the training data:
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
```

```python
# Validation:

# The NB model is used to make predictions on the test data:
predictions = nb_model.predict(x_test)

# The performance of the NB model is evaluated:
accuracy = accuracy_score(y_test, predictions)
print("Accuracy NB:",accuracy)

# Confusion Matrix
cm = skm.confusion_matrix(y_test, predictions)
disp = skm.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=None)
disp.plot()
plt.show()
```

5. Support vector machine (SVM):

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(x_train, y_train)

#In this model, support vectors are calculated from Lagrange Multipliers.
#These multipliers are solved with quadratic optimization, something that is very computationally expensive.
#The above translates into a longer training time.
```

```python
#Validation

predictions = svm_model.predict(x_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy SVC:",accuracy)

cm = skm.confusion_matrix(y_test, predictions)
disp = skm.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=None)
disp.plot()
plt.show()

#This model is one of the best.
```

6. K- nearest neighbor (K-NN):

```python
from sklearn.neighbors import KNeighborsClassifier

# Training:
# A K-NN model is created and fitted with the training data:
# K represents the number of neighbors that will be considered, it must be an odd value.

k=3
knn_model= KNeighborsClassifier(n_neighbors=k)
knn_model.fit(x_train, y_train)
```

```python
# Validation:

# The K-NN model is used to make prediction on the test data:
predictions = knn_model.predict(x_test)

# The performance of the K-NN model is evaluated:
accuracy = accuracy_score(y_test, predictions)
print("Accuracy K-NN:",accuracy)

# Confusion matrix:
cm = skm.confusion_matrix(y_test, predictions)
disp = skm.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=None)
disp.plot()
plt.show()
```
